```
import pandas as pd
import matplotlib.pyplot as plt
import os
import json
import seaborn as sns
import plotly.express as px
```

#1 What areas have the most influence on revenue?

```
def data_transfer(file):
    df = pd.read_csv(file)
    df['release_date'] = pd.to_datetime(df['release_date']).apply(lambda x: x.date())
    json_columns = ['genres','production_countries','production_companies']
    for col in json_columns:
        df[col] = df[col].apply(json.loads)
    return df

def get_json_object(col, range):
    output = col
    try:
        for i in range:
            output = output[i]
        return output
    except IndexError or KeyError:
        return pd.np.nan

df = data_transfer('tmdb_5000_movies.csv')
df['area'] = df['production_countries'].apply(lambda x: get_json_object(x, [0,'iso_3166_1']))
df1 = df[['revenue','area']]
df.columns
df.head(5)
```

```
print(df1.describe())
print(df1.shape)

#plot = px.bar(df1, x='area', y='revenue')
#plot.show()
fig = px.box(df1, x="area", y="revenue")
fig.show()
```

```
mean_inx = df1.groupby(['area'], as_index=False)['revenue'].median()
print(mean_inx)

plot1 = px.bar(mean_inx, x='area', y='revenue')
plot1.show()
```
